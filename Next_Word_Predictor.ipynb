```python
texts = texts.replace('“','')
texts = texts.replace('”','')
texts = texts.replace('‘','')
texts = texts.replace('’','')
texts = texts.replace('—','')
texts = texts.replace('…','')
texts = texts.replace('–','')
# texts = texts.replace('I','')
texts = texts.replace('II','')
texts = texts.replace('III','')
texts = texts.replace('IV','')
texts = texts.replace('V','')
texts = texts.replace('VI','')
texts = texts.replace('VII','')
texts = texts.replace('VIII','')
texts = texts.replace('IX','')
texts = texts.replace('X','')
texts = texts.replace('XI','')
texts = texts.replace('XII','')
texts = texts.replace('\ufeff','')
texts = texts.replace('\n\n','\n')
texts = texts.replace('\n\n\n','\n')
texts = texts.replace('\n\n\n\n','\n')
```

# Remove URLs

```python
print(texts)
```

```python
import re

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

texts = remove_url(texts)

print(texts)
```

# Remove Punctuation

```python
import string, time

string.punctuation
```

```python
exclude = """!"#$%&'()*+,-./:;<=>?@[]^_`{|}~"""

# def remove_punc(text):
#     for char in exclude:
#         text = text.replace(char,'')
    # return text

def remove_punc(text):
    return text.translate(str.maketrans('','',exclude))

texts = remove_punc(texts)

print(texts)
```

# Tokenization

There is various options for Tokenization
1. split
2. RegEx
3. NLTK
4. Spacy
5. Tensorflow
- We used Tensorflow for Tokenization

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([texts])
```

```python
len(tokenizer.word_index)
```

```python
tokenizer.word_index
```

```python
import pickle
with open('tokenizer.pkl', 'wb') as file:
    pickle.dump(tokenizer, file)
```

```python
input_sequences = []

for i in texts.split('\n'):
  # print(i)
  tokenized_sequence = tokenizer.texts_to_sequences([i])[0]

  # loop for creating dataset
  for j in range(1, len(tokenized_sequence)):
    input_sequences.append(tokenized_sequence[:j+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])

max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

padded_input_sequences = pad_sequences(input_sequences,maxlen=max_len,padding='pre')
```

```python
padded_input_sequences
```

```python
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
print(X.shape)
print(y.shape)
```

```python
len(tokenizer.word_index)
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=8488)
```

```python
y.shape
```

# Architecture

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,Dense,LSTM
```

```python
model = Sequential()

model.add(Embedding(input_dim = 8488, output_dim = 100, input_length=17))

model.add(LSTM(150,return_sequences=True))
model.add(LSTM(150))

model.add(Dense(8488, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X,y,epochs=40)
```

```python
import numpy as np
import time

text = 'i have'

for i in range(5):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]

  # padding
  pad_token_text = pad_sequences([token_text],maxlen=17,padding='pre')

  # predict
  pos = np.argmax(model.predict(pad_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + ' ' + word
      print(text)
      time.sleep(1)
```
